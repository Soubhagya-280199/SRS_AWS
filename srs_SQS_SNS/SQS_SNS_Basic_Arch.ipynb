AWS SQS & SNS:

AWS SQS (Simple Queue Service):

    ->AWS SQS is fully managed message queue service offered by AWS.

    ->It enables decoupling & scaling of microservices, distributed systems and serverless applications.
    

There are two types of Queues:

    Standard Queue : 

                    Offers Best effort ordering. More than one delivery at a time and nearly unlimited throughput[the amount of material or items passing through a system or process.] 

                    But there is chances we can see the a single message coming multiple times into system which are duplicate processing.

    FIFO Queue (First In First Out):
                    Gurantees the order in which messages are sent and processed exactly Once.

                    Most organization using this because of data processing in a sequence with no duplicacy.
